```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('Sample Images/pan card img.JPG')
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.imread('Sample Images/abhimanyu pfp.jpg', 0)
```

```python
# blur

# blurring is often used to reduce noise, and better recognize features

img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
figure_size = 7
result = cv2.blur(img, (figure_size, figure_size))
plt.figure(figsize=(11, 6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB)), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB)), plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
(height, width) = img1.shape[:2]
zoom = cv2.resize(img1, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom = cv2.resize(img1, (width*2, height*2), interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
m = cv2.getRotationMatrix2D((width/2, height/2), 30, 1)
rotate = cv2.warpAffine(img1, m, (width, height))
plt.imshow(rotate)
```

```python
# rotate img1 by 90 degrees

m = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
rotate = cv2.warpAffine(img1, m, (width, height))
plt.imshow(rotate)
```

```python
# translate img1 by 50 pixels to the right and 100 pixels down
arr = np.float32([[1, 0, 50], [0, 1, 100]])
translate = cv2.warpAffine(img1, arr, (width, height))
plt.imshow(translate)
```
